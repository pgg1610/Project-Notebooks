## Linear regression using PyTorch
1. Design model (input, output size, forward pass) 
2. Construct loss function and optimizer 
3. Training loop 
    - forward pass: compute prediction 
    - backward pass: gradients 
    - update weights using gradient descent 

```python
import torch 
import torch.nn as nn 
import numpy as np 
from sklearn import datasets 
import matplotlib.pyplot as plt 
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
```

```python
# (0) Prepare data
x_numpy, y_numpy = datasets.make_regression(n_samples = 100, n_features=1, noise=20, random_state=42)
type(x_numpy)
x = torch.tensor(x_numpy, dtype=torch.float32).view(-1,1)
y = torch.tensor(y_numpy, dtype=torch.float32).view(-1,1)
n_sample, n_features = x.shape
```

```python
# (1) model making
input_features = n_features #Num of dimensions of X (input feature space)
output_features = y.shape[-1] #Num of output dimensions 
model = nn.Linear(in_features=input_features, out_features=output_features)
```

```python
# (2) Loss and the optimizer
criterion = nn.MSELoss()
learning_rate = 0.01 
optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
```

```python
# (3) Training loop
num_epochs = 100
loss = []
for epoch in range(num_epochs):
    #forward pass 
    y_pred = model(x)
    
    #Evaluate loss function 
    loss_value = criterion(y_pred, y)
    
    #Backward pass 
    loss_value.backward() #FUCK YEAH! Does the backprop for you! 
    
    #Update weights 
    optimizer.step()
    
    #Flush the gradients since it to auto-additive 
    optimizer.zero_grad()
    loss.append(loss_value.item())
    if epoch % 200 == 0: 
        [w, b] = model.parameters() 
        print(f'epoch: {epoch+1}, loss = {loss_value.item():.4f}')
```

```python
#Predicted y -- detach the grad first 
predicted_y = model(x).detach().numpy() #Detach the gradient operator from the computation graph 

#Plotting 
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,8))
ax1.plot(range(num_epochs), loss)
ax2.scatter(y_numpy, predicted_y)
ax3.scatter(x_numpy, y_numpy);
```
