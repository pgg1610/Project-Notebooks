```python
import matplotlib.pyplot as plt
import numpy as np

from sklearn.gaussian_process import GaussianProcessRegressor as GPR
from sklearn.gaussian_process import kernels

import demo
import pybo
```

```python
EI = pybo.acquisition.ExpectedImprovement(delta = 0.01)
LCB = pybo.acquisition.LowerConfidenceBound(sigma = 1.96)
```

## Defining the params of the GPR

```python
constant = kernels.ConstantKernel()
matern = kernels.Matern(nu = 2.5)
rbf = kernels.RBF()
```

# A Two-Dimensional Example -- Rosenbrock

```python
objective = pybo.objectives.rosenbrock
domain = np.array([[-2.0, -2.0], [2.0, 2.0]])

# Define domain
xx_pts = np.linspace(domain[0,0], domain[1,0], 50) 
yy_pts = np.linspace(domain[0,1], domain[1,1], 50)
xy_pts = np.c_[[z.flatten() for z in np.meshgrid(xx_pts, yy_pts)]].T

#Calculate the function value 
f_grid = objective(xy_pts).reshape((yy_pts.size, xx_pts.size)) 

#Generate random numbers from a uniform distribution -- 5 points 
num_points = 20
generator = np.random.default_rng(42)
X0_mb = generator.uniform(domain[0,:], domain[1,:], size = (num_points, 2))
y0_mb = objective(X0_mb)
```

```python
gpr_mb = GPR(kernel = constant * rbf, alpha = 1e-3, n_restarts_optimizer = 20, normalize_y = False, random_state = 42)
gpr_mb.fit(X0_mb, y0_mb)

(m0, s0) = gpr_mb.predict(xy_pts, return_std = True)

m0 = m0.flatten()
gpr_mb.kernel_
```

## Illustrate the Initial Sample

```python
(fig_mb, ax_mb) = demo.illustrate_2d_gpr(objective, gpr_mb, xx_pts, yy_pts, LCB)
```

## Run a Few Iterations and Assess

```python
pkwargs = {"num_sample": 100,
           "num_improve": 10,
           "generator": generator,
          "return_data":True}

# num_sample -- Number of random initial conditions to generate when searching for an optimizer of the acquisition function. (What is this?)

res_mb, each_iter = pybo.bayesian_optimization(objective, gpr_mb, LCB, domain, prop_kwargs = pkwargs)
gpr_mb.fit(res_mb["X"], res_mb["y"]) # Incorporate final point into plots.
(fig_mb, ax_mb) = demo.illustrate_2d_gpr(objective, gpr_mb, xx_pts, yy_pts, LCB)
```
