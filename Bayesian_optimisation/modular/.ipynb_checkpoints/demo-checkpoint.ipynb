```python
import matplotlib.pyplot as plt
import numpy as np
%config InlineBackend.figure_format = 'retina'

from sklearn.gaussian_process import GaussianProcessRegressor as GPR
from sklearn.gaussian_process import kernels

import demo_plotting
import scripts
```

```python
plot_params = {
'font.size' : 22,
'axes.titlesize' : 24,
'axes.labelsize' : 20,
'axes.labelweight' : 'bold',
#'lines.linewidth' : 3,
#'lines.markersize' : 10,
'xtick.labelsize' : 16,
'ytick.labelsize' : 16,
}
 
plt.rcParams.update(plot_params)
```

## ## Import the acquisition functions implemented 

```python
EI = scripts.acquisition.ExpectedImprovement(delta = 0.01)
LCB = scripts.acquisition.LowerConfidenceBound(sigma = 1.96)
```

# A One-Dimensional Example

### ### Egg-carton objective function 

```python
objective = scripts.objectives.egg_carton
(low, high) = (0.0, 10.0)
domain = np.array([[low], [high]])
x_pts = np.linspace(low, high, 1000).reshape(-1, 1)
y_pts = objective(x_pts)
```

```python
num_sample_points = 10
noise_ = 0.1
generator = np.random.default_rng(42)
x_sample = generator.uniform(low, high, size = (num_sample_points, 1))
y_sample = objective(x_sample, noise_)
```

```python
fig, ax = plt.subplots(1,1, figsize=(8,3))
ax.plot(x_pts, y_pts, 'k-', linewidth=2.0, label='Ground truth')
ax.plot(x_sample, y_sample, 'ro', label='Noisy sampled points')

ax.set_xlabel(r"$x$", fontsize = 20)
ax.set_ylabel(r"$f(x)$", fontsize = 20)
ax.set_title("Initial setup", fontsize = 20)
ax.legend(fontsize = 15)
ax.grid(True)
```

## ## Fit a GPR model (surrogate function) to the sampled points

```python
constant = kernels.ConstantKernel()
matern = kernels.Matern(nu = 2.5)
rbf = kernels.RBF()

gpr_model = GPR(kernel = constant*rbf, alpha = 1e-3, n_restarts_optimizer = 20, normalize_y = False, random_state = 42)
gpr_model.fit(x_sample, y_sample)

(mean_pred, stddev_pred) = gpr_model.predict(x_pts, return_std = True)
gpr_model.kernel_
```

## Illustrate the Initial Sample

```python
(fig_ec, ax_ec) = demo_plotting.illustrate_1d_gpr(objective, gpr_model, x_pts, EI, LCB)
```

## Run a Few Iterations and Assess

```python
pkwargs = {"num_sample": 50,
           "num_improve": 5,
           "generator": generator}

res_ec, _ = scripts.bayesian_optimization(objective, gpr_model, LCB, domain, max_iter=10, noise=noise_, prop_kwargs = pkwargs)
gpr_model.fit(res_ec["X"], res_ec["y"]) # Incorporate final point into plots.
```

```python
(fig_ec, ax_ec) = demo_plotting.illustrate_1d_gpr(objective, gpr_model, x_pts, EI, LCB, num_sample_points)
```

```python
res_ec, _ = scripts.bayesian_optimization(objective, gpr_model, LCB, domain, noise=noise_, prop_kwargs = pkwargs)
gpr_model.fit(res_ec["X"], res_ec["y"]) # Incorporate final point into plots.
```
