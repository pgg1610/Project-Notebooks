## Rosenbrack function

$$ f(x,y) = (1-x)^2 - 100(y-x^2)^2 $$

```python
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as colors
import plotly.express as px
from matplotlib import cm
import seaborn as sns 

sns.set(style="white", color_codes=True)
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
```

```python
X = np.arange(-2, 3, 0.001).reshape(-1,1)
Y = np.arange(-2, 3, 0.001).reshape(-1,1)
(XX, YY) = np.meshgrid(X, Y, sparse=True)

def rosenbrack(x,y):
    f = (1 - x)**2 - 100 * (y - x**2)**2
    return f
```

```python
print(XX.shape, YY.shape)
```

```python
Z = rosenbrack(XX, YY)
```

```python
index_array = np.random.choice(len(X), 100, replace=False)
X_known, Y_known = X[index_array], Y[index_array]
Z_known = rosenbrack(X_known, Y_known)

fig = plt.figure(figsize=(10,8))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(X_known, Y_known, -Z_known, c='r', marker='o', s=100, depthshade=False)
ax.plot_surface(XX, YY, -Z, cmap=cm.viridis, linewidth=0, alpha=0.5, antialiased=False, norm=colors.PowerNorm(gamma=0.2))
```
